## Dataset
The dataset used is `student-mat.csv`, which contains student information and academic records for Mathematics course students.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Analysis Question
Average Final Grade(G3)

```python
average_grade = df["G3"].mean()

print("Average Final Grade:", average_grade)
```

Student scoring above 15

```python
high_scorers = df[df["G3"]>15]
print("Number of students scoring above 15:", len(high_scorers))
```

Is study time correlated with performance?

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation:",correlation)
```

Which gender performs better on average?

```python
gender_performance = df.groupby("sex")["G3"].mean()
print(gender_performance)
```

#Visualization
